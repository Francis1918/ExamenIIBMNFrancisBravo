## Comparación con la solución real
La solución real de la ecuación diferencial es:
$y(t)=t^2 +1/3e^{-5t}$

* Grafique la solución real y la aproximación obtenida con el método de Euler en una misma gráfica.



```python
import matplotlib.pyplot as plt
import numpy as np


def f(t, y):
    return -5 * y + 5 * t**2 + 2 * t

a, b = 0, 1  
y_t0 = 1/3   
N = 10      


euler_ys, euler_ts, h = ODE_euler(a=a, b=b, f=f, y_t0=y_t0, N=N)

def exact_solution(t):
    return t**2 + (1/3) * np.exp(-5*t)


t_exact = np.linspace(a, b, 100)
y_exact = exact_solution(t_exact)

# Graficar la solución aproximada y exacta
plt.figure(figsize=(8, 5))
plt.plot(t_exact, y_exact, label="Solución Exacta", color='g')
plt.scatter(euler_ts, euler_ys, color='r', label="Euler", marker='o')
plt.plot(euler_ts, euler_ys, linestyle='dashed', color='r')
plt.xlabel("t")
plt.ylabel("y")
plt.title("Método de Euler vs Solución Exacta")
plt.legend()
plt.grid()
plt.show()
```

# Errores
Calcule el error relativo para diferentes valores de N.

```python
# Definir la función de error relativo
def relative_error(exact, approx):
    return abs((exact - approx) / exact)
```

### N = 10

```python

# Resolver la EDO con el método de Euler
euler_ys, euler_ts, h = ODE_euler(a=a, b=b, f=f, y_t0=y_t0, N=N)

# Calcular errores relativos en cada punto
errors = [relative_error(exact_solution(t), y_euler) for t, y_euler in zip(euler_ts, euler_ys)]

# Calcular el promedio del error relativo
average_relative_error = sum(errors) / len(errors)

# Mostrar resultado en notación científica
average_relative_error, "{:.2e}".format(average_relative_error)
```
