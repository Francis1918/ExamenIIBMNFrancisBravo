# Splines cÃºbicos

```python
def Spline(x: float, x0: float, pars: dict[str, float]) -> float:
    a = pars["a"]
    b = pars["b"]
    c = pars["c"]
    d = pars["d"]
    return a + b * (x - x0) + c * (x - x0) ** 2 + d * (x - x0) ** 3
```

```python
import matplotlib.pyplot as plt
import numpy as np

xs = [-1, 0, 1]
ys = [1, 5, 3]
s = [
    {"a": 1, "b": 5.5, "c": 0, "d": -1.5},
    {"a": 5, "b": 1, "c": -4.5, "d": 1.5},
]
for i, x_i in enumerate(xs[:-1]):
    _x = np.linspace(x_i, xs[i + 1], 20)
    _y = Spline(_x, x_i, s[i])
    plt.plot(_x, _y, color="red")

plt.scatter(xs, ys)
plt.xlabel("x")
plt.ylabel("y")
plt.title("InterpolaciÃ³n con splines cÃºbicos")
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# 1) Nodos y pendiente deseada
x0, y0 = -1, 1
x1, y1 =  0, 5
x2, y2 =  1, 3
m = -3

# 2) Coeficientes (hallados analÃ­ticamente)
a0, b0, c0, d0 = 1,  7.5,  0,  -3.5
a1, b1, c1, d1 = 5, -3.0, 1.5, -0.5

# 3) Definir funciones spline
def S0(x):
    return a0 + b0*(x - x0) + c0*(x - x0)**2 + d0*(x - x0)**3

def S1(x):
    return a1 + b1*(x - x1) + c1*(x - x1)**2 + d1*(x - x1)**3

# 4) Graficar
xs0 = np.linspace(x0, x1, 100)
xs1 = np.linspace(x1, x2, 100)

plt.figure()
plt.plot(xs0, S0(xs0), 'C0', label='S0 en [â1,0]')
plt.plot(xs1, S1(xs1), 'C1', label='S1 en [0,1]')
plt.scatter([x0, x1, x2], [y0, y1, y2], color='k', zorder=5)

# lÃ­nea de pendiente m en x1
x_line = np.linspace(x1-0.5, x1+0.5, 10)
y_line = y1 + m*(x_line - x1)
plt.plot(x_line, y_line, 'k--', label=f'Pendiente m={m}')

plt.legend()
plt.xlabel('x')
plt.ylabel('y')
plt.title('Spline clamped con pendiente m en xâ')
plt.grid(True)
plt.show()
```
